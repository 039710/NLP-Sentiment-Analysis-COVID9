# Opening Dataset and convert it into Array

```python
dataset = []
with open ('dataset/datasetfix1.csv',encoding="latin1") as csvfile:
    readCSV = csv.DictReader(csvfile, delimiter=';')
    
    for row in readCSV:
        tweets = row['tweet']    
        usernames = row['username']
        hashtags = row['hashtag']
        dates = row['date']
        intent = row['intent']
        emotion = row['emotion']
        dataset.append({
            'date' : dates,
            'username' : usernames,
            'tweet' : tweets,
            'hashtag' : hashtags,
            'intent' : intent,
            'emotion' : emotion
        })
            
        
print("total tweets : " , len(dataset))
```

The purpose of language processing are to remove non Indonesian word such as regional words, english words, etc.

The codes below show the proses of loading list of english words,typo words, abbreviations, and slang words, that already listed manually by filetring each unique words in the document.

## Loading list of English words

```python
englishWords = []
#start english words
def getEnglishWords(englishWordsFileName):
    #read the englishWords file and build a list
    englishWords = []
    fp = open(englishWordsFileName, 'r',encoding ='latin1')
    line = fp.readline()
    while line:
        word = line.strip()
        englishWords.append(word.lower())
        line = fp.readline()
    fp.close()
    return englishWords
st = open('feature_list/englishwords.txt', 'r',encoding = 'latin1')
englishWords = getEnglishWords('feature_list/englishwords.txt')
```

```python
#check if word contain english words
for n in range(len(word)):
    if word[n] not in englishWords:
        filtered_word.append(word[n])
print('Total word without english word',len(filtered_word)) 
filtered_word.sort()
```

## Load list of typo words

```python
typo_list = []
with open ('feature_list/typo.csv',encoding="latin1") as csvfile:
    readCSV = csv.DictReader(csvfile, delimiter=',')
    for row in readCSV:
        word = row['typo']    
        standard_word = row['standard_form']
        typo_list.append({
            'word' : word,
            'standard_word' : standard_word
        })  
print("total typo words : " , len(typo_list))
```

## Load list of abbreviations words

```python
abbreviation_list = []
with open ('feature_list/abbreviation.csv',encoding="latin1") as csvfile:
    readCSV = csv.DictReader(csvfile, delimiter=',')
    for row in readCSV:
        word = row['abbreviations']    
        standard_word = row['non-abbreviations']
        abbreviation_list.append({
            'word' : word,
            'standard_word' : standard_word
        })      
print("total abbreviation words : " , len(abbreviation_list))
```

## Load list of slang words

```python
slang_list = []
with open ('feature_list/slangwords.csv',encoding="latin1") as csvfile:
    readCSV = csv.DictReader(csvfile, delimiter=',')
    for row in readCSV:
        word = row['slang_word']    
        standard_word = row['formal_word']
        slang_list.append({
            'word' : word,
            'standard_word' : standard_word
        })     
print("total slang words : " , len(slang_list))
```

## Load combination list of slang words,abbreviation words, and typo words

```python
word_list = []
with open ('feature_list/word_list.csv',encoding="latin1") as csvfile:
    readCSV = csv.DictReader(csvfile, delimiter=',')
    for row in readCSV:
        word = row['word']    
        standard_word = row['formal_word']
        word_list.append({
            'word' : word,
            'standard_word' : standard_word
        })    
print("total word_list : " , len(word_list))
```

## Language Processing

```python
def languagePreprocessing(dataset):
    for row in dataset:
        word_tokens = word_tokenize(row['tweet']) 
        filtered_word = []
        for x in range(len(word_tokens)):
            for n in range(len(word_list)):
                if word_tokens[x] == word_list[n]['word']:
                    word_tokens[x] = word_list[n]['standard_word']
                    filtered_word.append(word_tokens[x])
            filtered_word.append(word_tokens[x])
            filtered_word = list(dict.fromkeys(filtered_word))
        row['tweet']= " ".join(filtered_word)
languagePreprocessing(dataset)
```

"After each list for each category listed, then I manually combine them into one csv file with two columns which are "word" and standard word".
By iterating each words after tokenization, if the word match in the document, then the word will be changed based on standard word in csv file.

# Stemming, Tokenization, Stopwords removal, remove unnecessary attribute

## Load indonesian stopwords list

```python
#Stopword
stopWords = []
#start getStopWordList
def getStopWordList(stopWordListFileName):
    #read the stopwords file and build a list
    stopWords = []
    stopWords.append('atuser')
    stopWords.append('url')

    fp = open(stopWordListFileName, 'r')
    line = fp.readline()
    while line:
        word = line.strip()
        stopWords.append(word)
        line = fp.readline()
    fp.close()
    return stopWords

st = open('feature_list/stopwordsID.txt', 'r')
stopWords = getStopWordList('feature_list/stopwordsID.txt')
#end
```

```python
def preprocessing(dataset):
    # create stemmer
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    count = 0
    for row in dataset:
        #print("before : ",row['tweet'])
        #convert sentence in lowercase
        row['tweet'] = row['tweet'].lower()
        row['tweet'] = re.sub('[^A-Za-z0-9 ]+','', row['tweet'])
        # remove number
        row['tweet'] = re.sub('[0-9]+', '', row['tweet'])
        row['tweet'] = re.sub(r'\d+', '', row['tweet'])
        #Convert www.* or https?://* to URL
        row['tweet'] = re.sub('((www\.[^\s]+)|(https?://[^\s]+))','URL',row['tweet'])
        #remove atuser words
        row['tweet'] = row['tweet'].replace('atuser','')
        #remove cc words
        row['tweet'] = row['tweet'].replace('cc','')
        #remove via words
        row['tweet'] = row['tweet'].replace('via','')
        # remove all single characters
        row['tweet'] = re.sub(r'\s+[a-zA-Z]\s+', ' ', row['tweet'])
        # Remove single characters from the start
        row['tweet'] = re.sub(r'\^[a-zA-Z]\s+', ' ', row['tweet']) 
        #Tokenize
        word_tokens = word_tokenize(row['tweet']) 
        #Stop word removal
        filtered_sentence = [w for w in word_tokens if not w in stopWords]
        row['tweet'] = " ".join(filtered_sentence)
        #stemming
        stop_sentence = [] 
        for v in filtered_sentence:
            stop_sentence.append(stemmer.stem(v))
        row['tweet'] = " ".join(stop_sentence)
        #print("after : ",row['tweet'])
preprocessing(dataset)
```

## Feature Extraction TF-IDF Random under sampling

```python
from imblearn.over_sampling import BorderlineSMOTE, SMOTE, ADASYN, SMOTENC, RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
X = df['tweet']
y = df['intent']
#TFIDF
tfidf_vectorizer = TfidfVectorizer() 
x_tfidf = tfidf_vectorizer.fit_transform(X) 
#Balancing dataset with SMOTE
oversample = RandomUnderSampler(random_state=42)
X , y  = oversample.fit_resample(x_tfidf, y) # X in this case already use tf-idf

#print(X)
print(Counter(y))
```
