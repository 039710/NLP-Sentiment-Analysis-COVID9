```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import csv
import re
import json
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from wordcloud import WordCloud
import seaborn as sn
import sklearn.metrics as metrics
from nltk.tokenize import sent_tokenize, word_tokenize
from sklearn import datasets, linear_model
from sklearn import preprocessing
from sklearn.preprocessing import PolynomialFeatures
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import word_tokenize, RegexpTokenizer
from sklearn.model_selection import train_test_split
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

# Opening Dataset and convert it into Array

```python
dataset = []
with open ('dataset/datasetfix1.csv',encoding="latin1") as csvfile:
    readCSV = csv.DictReader(csvfile, delimiter=';')
    
    for row in readCSV:
        tweets = row['tweet']    
        usernames = row['username']
        hashtags = row['hashtag']
        dates = row['date']
        intent = row['intent']
        emotion = row['emotion']
        dataset.append({
            'date' : dates,
            'username' : usernames,
            'tweet' : tweets,
            'hashtag' : hashtags,
            'intent' : intent,
            'emotion' : emotion
        })
            
        
print("total tweets : " , len(dataset))
```

# Data Preprocessing

## Loading stop word list

```python
#Stopword
stopWords = []
#start getStopWordList
def getStopWordList(stopWordListFileName):
    #read the stopwords file and build a list
    stopWords = []
    stopWords.append('atuser')
    stopWords.append('url')

    fp = open(stopWordListFileName, 'r')
    line = fp.readline()
    while line:
        word = line.strip()
        stopWords.append(word)
        line = fp.readline()
    fp.close()
    return stopWords

st = open('feature_list/stopwordsID.txt', 'r')
stopWords = getStopWordList('feature_list/stopwordsID.txt')
#end
```

# Stemming, Tokenization, Stopwords removal, remove unnecessary attribute

```python
# negation handling
```

```python
## Rumusan masalah
# perbandingan performa classifier
# data imbalance methode ny apa aja, trs kenapa di project ini yg di pake metode ini .
# -> negation handling -> tidak bagus = jelek 
# perbandingan menggunakan N-Gram sama tidak menggunakan N-Gram

## apakah  n gram dapat mempengaruhi nilai performa klasifikasi dalam bahasa indonesia.


# pos tagging
```

```python
def preprocessing(dataset):
    # create stemmer
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    count = 0
    for row in dataset:
        #convert sentence in lowercase
        row['tweet'] = row['tweet'].lower()
        row['tweet'] = re.sub('[^A-Za-z0-9 ]+','', row['tweet'])
        #Convert www.* or https?://* to URL
        row['tweet'] = re.sub('((www\.[^\s]+)|(https?://[^\s]+))','URL',row['tweet'])
        #remove atuser words
        row['tweet'] = row['tweet'].replace('atuser','')
        #Tokenize
        word_tokens = word_tokenize(row['tweet']) 
        #Stop word removal
        filtered_sentence = [w for w in word_tokens if not w in stopWords]
        row['tweet'] = " ".join(filtered_sentence)
        #stemming
        stop_sentence = [] 
        for v in filtered_sentence:
            stop_sentence.append(stemmer.stem(v))
        row['tweet'] = " ".join(stop_sentence)
preprocessing(dataset)
```

# Split dataset to datatrain and datatest

```python
import imblearn
from collections import Counter
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE
from matplotlib import pyplot
from numpy import where

X = df['tweet']
y = df['intent']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, random_state=101)

# Bag of Word transformation
bow_transformer = CountVectorizer().fit(X)
X = bow_transformer.transform(X)

counter = Counter(y)
print(counter)
```

```python
print("Size of test data : " ,len(y_train), "," , len(y_test))
```

# Implementing SMOTE

```python

# Balancing positive and negative class
oversample = SMOTE()
X, y = oversample.fit_resample(X, y)
print(Counter(y))
```

# Naive Bayes

```python
nb = MultinomialNB()
clf = nb.fit(X_train, y_train)
y_preds = nb.predict(X_test)
print (metrics.classification_report(y_test, y_preds))
```

```python
cm = metrics.confusion_matrix(y_test,y_preds)

sn.heatmap(cm, annot=True,  fmt='.2f', xticklabels = ["Negative", "Positive"] , yticklabels = ["Negative", "Positive"],)
plt.ylabel('True label',fontsize=12)
plt.xlabel('Predicted label',fontsize=12)
```

# Decision Tree

```python

clf = DecisionTreeClassifier().fit(X_train, y_train)
y_preds = clf.predict(X_test)
print (metrics.classification_report(y_test, y_preds))
```

```python
cm = metrics.confusion_matrix(y_test,y_preds)

sn.heatmap(cm, annot=True,  fmt='.2f', xticklabels = ["Negative", "Positive"] , yticklabels = ["Negative", "Positiev"],)
plt.ylabel('True label',fontsize=12)
plt.xlabel('Predicted label',fontsize=12)
```

# Logistic Regression

```python
cm = metrics.confusion_matrix(y_test,y_preds)
sn.heatmap(cm, annot=True,  fmt='.2f', xticklabels = ["Negative", "Positive"] , yticklabels = ["Negative", "Positive"],)
plt.ylabel('True label',fontsize=12)
plt.xlabel('Predicted label',fontsize=12)
```
